#### **2. Écrire une fonction prenant comme argument un entier n et simulant n pas de marche aléatoire puis affichant la position finale. Modifier le programme pour qu’il affiche la liste de toutes les positions intermédiaires et les représente graphiquement. Vous afficherez en abscisse le numéro de l’étape et en ordonnée la position de l’objet à cette étape. Tester votre programme avec différentes valeurs de n**

```python
import matplotlib.pyplot as plt
import numpy as np

def random_etape():
    return np.random.choice([-1, 1])

def marche_aleatoire(n):
    positions = [0]
    for _ in range(n):
        etape = random_etape()
        positions.append(positions[-1] + etape)

    print(f"Position finale après {n} étapes: {positions[-1]}")
    print(f"Positions intermédiaires: {positions}")

    # Plot the positions
    plt.plot(positions)
    plt.xlabel("Numéro de l'etape")
    plt.ylabel('Position')
    plt.title(f'Marche aléatoire de {n} étapes')
    plt.show()

values = [10**2, 10**5]
for n in values:
    print(f"Simulation pour n = {n}")
    marche_aleatoire(n)
```
